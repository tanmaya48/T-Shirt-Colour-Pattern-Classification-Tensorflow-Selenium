```python
import pandas as pd

df = pd.read_csv('shirts.csv')

df.head()
```

```python
data = []
labels = []

from tensorflow.keras.preprocessing.image import img_to_array
import cv2

IX = 32
IY = 32

for name in df.Id:
    
    image = cv2.imread('shirts/'+str(name))
    image = cv2.resize(image, (IX,IY) )
    image = img_to_array(image)
    data.append(image)
    
    
for ind in df.index: 
    label = [ df['Color'][ind], df['Pattern'][ind] ] 
    labels.append(label)
    
```

```python
import numpy as np

data = np.array(data, dtype="float") / 255.0
labels = np.array(labels)

print(labels)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
labels = mlb.fit_transform(labels)

print(mlb.classes_)
print(labels[0])
    
    
```

```python
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D

inputShape = (IX, IY, 3)


model = Sequential()

model.add(Conv2D(32, (3, 3), padding="same",input_shape=inputShape))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Flatten()) 

model.add(Dense(64))
model.add(Activation('sigmoid'))


out = len(mlb.classes_)

model.add(Dense(out))
model.add(Activation('sigmoid'))
                    
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])          
          
          
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split


(trainX, testX, trainY, testY) = train_test_split(data,labels, test_size=0.2, random_state=42)



IDG = ImageDataGenerator(rotation_range=25, width_shift_range=0.1
                         ,height_shift_range=0.1, shear_range=0.2, 
                         zoom_range=0.2,horizontal_flip=True,vertical_flip=True, fill_mode="nearest")
```

```python
batch = 32
E = 50


model.fit(x=IDG.flow(trainX, trainY, batch_size= batch),
          epochs=E ,verbose=1)
```

```python
image = cv2.imread('test_shirt.jpeg')
image = cv2.resize(image, (IX,IY) )
image = img_to_array(image)

test_img = []

test_img.append(image) 

test_img = np.array(test_img, dtype="float") / 255.0
```

```python
pred = model.predict(test_img)

print(pred)

for i in range(0,len(pred[0])):
    if pred[0,i] > 0.5:
        print(mlb.classes_[i])
```
